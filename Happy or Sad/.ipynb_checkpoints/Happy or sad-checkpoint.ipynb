```python
import tensorflow as tf
import os
import zipfile
```

```python
zip_file_address = "happy-or-sad.zip"
!wget --no-check-certificate \
    "https://storage.googleapis.com/laurencemoroney-blog.appspot.com/happy-or-sad.zip" \
    -O zip_file_address
zip_ref = zipfile.ZipFile(zip_file_address, 'r')
zip_ref.extractall(zip_file_address[0:-17] + '/h-or-s')
zip_ref.close()
```

```python
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

class my_Callback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('acc')>0.999:
            print('\n Reached 99.9 % of accuracy so canceling training')
            self.model.stop_training = True
    
callbacks = my_Callback()
```

```python
import os

happy_dir_list = os.listdir(zip_file_address[0:-17] + '/h-or-s/' + '/happy/')
sad_dir_list = os.listdir(zip_file_address[0:-17] + '/h-or-s/' + '/sad/')
print(len(happy_dir_list), len(sad_dir_list))
```

```python
from random import randint
plt.figure(figsize = (10, 10))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(mpimg.imread(zip_file_address[0:-17] + '/h-or-s/' + '/happy/' + 
                            happy_dir_list[randint(0, len(happy_dir_list) - 1)]))
plt.tight_layout()
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16, (3, 3), input_shape=(150, 150, 3), 
                                                           activation = 'relu'),
                                    tf.keras.layers.MaxPool2D((2,2)),
                                    
                                    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D((2,2)),
                                    
                                    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D((2, 2)),
                                    
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(512, activation = 'relu'),
                                    tf.keras.layers.Dense(1, activation = 'sigmoid')
    
])
model.summary()
```

```python
from keras.optimizers import RMSprop
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_data_generator = ImageDataGenerator(rescale=1/255)
train_generator = train_data_generator.flow_from_directory(zip_file_address[0:-17] + '/h-or-s/',
                                                           target_size = (150, 150),
                                                          batch_size = 10, 
                                                           class_mode = 'binary',)

                                                          # save_to_dir= zip_file_address[0:-17] + '/images/', 
                                                          # save_prefix='aug_shifted', 
                                                          # save_format='png'
                                                          # )
```

```python
model.fit_generator(train_generator, steps_per_epoch=20, epochs = 30, verbose=1, callbacks=[callbacks])
```

```python
from keras.preprocessing import image
import glob

files = glob.glob(zip_file_address[0:-17] + '/test_happy_or_sad/*')

plt.figure(figsize = (10, 10))
for file, i in zip(files, range(len(files))):
    plt.subplot(4, 4, i + 1)

    test_dir_list = os.listdir(zip_file_address[0:-17] + '/test_happy_or_sad/')
    test_image_path = zip_file_address[0:-17] + '/test_happy_or_sad/' + test_dir_list[i + 1]
    plt.imshow(mpimg.imread(test_image_path))

    test_img = image.load_img(test_image_path, target_size = (150, 150))
    test_img_array = image.img_to_array(test_img)
    test_img_arr_expand = np.expand_dims(test_img_array, axis = 0)
    test_img_arr_expand_vstack = np.vstack([test_img_arr_expand])
    classes = model.predict(test_img_arr_expand_vstack, batch_size=2)
    if classes[0][0] == 1.0:
        plt.title('Happy :)')
    else:
        plt.title('Sad :(')
        
plt.tight_layout()
classes
```
