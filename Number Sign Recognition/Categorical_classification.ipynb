```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
```

```python
x_l = np.load('X.npy')
Y_l = np.load('Y.npy')
```

```python
%matplotlib notebook
img_size = 64
plt.subplot(1, 2, 1)
plt.imshow(x_l[0,:, :].reshape(img_size, img_size))
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(x_l[900].reshape(img_size, img_size))
plt.axis('off')
```

```python
X = x_l
Y = Y_l
print(x_l.shape)
print(Y_l.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 42)
print(X_train.shape[0])
print(X_test.shape[0])
```

```python
X_train_flatten = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_test_flatten = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])
print(X_train_flatten.shape)
print(X_test_flatten.shape)
```

```python
x_train = X_train_flatten.T
x_test = X_test_flatten.T
y_test = Y_test.T
y_train = Y_train.T
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
col_index = {0: 9, 1:0, 2:7, 3:6, 4:1, 5:8, 6:4, 7:3, 8:2, 9:5}
plt.figure(figsize=(8, 8))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.imshow(X_train[i].reshape(img_size, img_size))
    plt.title('Number: {}'.format(col_index[Y_train[i].argmax()]), fontsize = 8)
plt.tight_layout()
```

# Implementing with Keras Libraries

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential 
from keras.layers import Dense, Flatten 
```

```python
x_train, y_train, x_test, y_test = x_train.T, y_train.T, x_test.T, y_test.T
model = Sequential() # initialize neural network
model.add(Dense(units = 128, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_train.shape[1]))
model.add(Dense(units = 20, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 200)
```

```python
x_train.shape[1]
```

```python
y_prediction = np.zeros((x_test.shape[0], 1)) # or np.zeros(y_test.shape)
prediction = model.predict(x_test)
for i in range(model.predict(x_test).shape[0]):
    if prediction[i, 0] <= 0.5:
        prediction[i, 0] = 0
    if prediction[i, 0] > 0.5:
        prediction[i, 0] = 1 
        
```

```python
print('Accuracy on test data')
print(model.evaluate(x_test, y_test)[1])
```

```python
(100 - 100 * np.mean(np.abs(y_test - prediction))) 
```

```python
%matplotlib notebook
plt.figure(figsize = (8, 8))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(x_test[i].reshape(img_size, img_size))
    if col_index[y_test[i].argmax()] == col_index[prediction[i].argmax()]:
        color = 'k'
    else:
        color = 'r'
        
    plt.title("Label: {}".format(col_index[y_test[i].argmax()]) + '\n' + 
              'Predicted label: {}'.format(col_index[prediction[i].argmax()]), color = color)
    
plt.tight_layout()
```
